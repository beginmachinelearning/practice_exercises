```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
from scipy import stats
from scipy.stats import norm
```

```python
all_data=pd.read_csv("data/all_data.csv")
```

```python
sns.distplot(all_data['Height'])
```

```python
sns.distplot(all_data['Weight'])
```

```python
sns.jointplot(all_data['Height'], all_data['Weight'])
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(all_data['Height'], all_data['Weight'], test_size=0.3, random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train.reshape(-1, 1), y_train)
```

```python
y_pred=lr.predict(X_test.reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y_test, y_pred))
rms
```
