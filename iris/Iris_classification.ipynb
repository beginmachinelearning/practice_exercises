```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
#### Data Explanation
#### ---------------------------
# 1. sepal length in cm
#    2. sepal width in cm
#    3. petal length in cm
#    4. petal width in cm
#    5. class: 
#       -- Iris Setosa
#       -- Iris Versicolour
#       -- Iris Virginica
# ---------------------------
```

```python
iris_data=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", delimiter=',', header=None)
```

```python
iris_data.columns=['sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'flower']
iris_data_base=iris_data
```

```python
iris_data
```

```python
plt.figure("Sepal Plots")
sns.regplot(iris_data.loc[iris_data.flower == 'Iris-setosa', 'sepallength'], iris_data.loc[iris_data.flower == 'Iris-setosa', 'sepalwidth'], label='Iris-setosa', color="r", fit_reg=False)
sns.regplot(iris_data.loc[iris_data.flower == 'Iris-versicolor', 'sepallength'], iris_data.loc[iris_data.flower == 'Iris-versicolor', 'sepalwidth'], label='Iris-versicolor', color="b", fit_reg=False)
sns.regplot(iris_data.loc[iris_data.flower == 'Iris-virginica', 'sepallength'], iris_data.loc[iris_data.flower == 'Iris-virginica', 'sepalwidth'], label='Iris-virginica', color="y", fit_reg=False)
plt.legend()
```

```python
### Iris setosa has a clear differentiation from the other two flower in terms of its sepal width and length
## Let us try to plot the petal length and width for the other 2
```

```python
plt.figure("Petal Plots")

sns.regplot(iris_data.loc[iris_data.flower == 'Iris-versicolor', 'petalwidth'], iris_data.loc[iris_data.flower == 'Iris-versicolor', 'sepalwidth'], label='Iris-versicolor', color="b", fit_reg=False)
sns.regplot(iris_data.loc[iris_data.flower == 'Iris-virginica', 'petalwidth'], iris_data.loc[iris_data.flower == 'Iris-virginica', 'sepalwidth'], label='Iris-virginica', color="y", fit_reg=False)
plt.legend()
```

```python
## Brilliant, these 2 have a clear difference in terms of their petal length. A classification algorithm should be able to segregate them easily
```

```python
y=iris_data_base['flower']
del iris_data['flower']
X=iris_data
```

```python
## Since we need numbers, let us label encode all the target variables to convert them to numbers
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
y=enc.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, train_size=0.2)
```

```python
X_train
```

```python
## Let us use different classifiers to make a prediction
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred=knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
