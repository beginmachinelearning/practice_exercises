```python
import random
    
def replace_with_proportion(frame, column):
    isnull = frame[column].isnull()
    sample = frame[column].dropna()
    sample=set(sample)
    print sample
    sample=set(sample)
    countframe=pd.DataFrame(columns=["name", "count", "proportion"])

    for x in sample:
        countframe.loc[-1] = [x,frame.loc[frame[column] == x].shape[0], 0 ]
        countframe.index = countframe.index + 1
    countframe['proportion']=countframe['count']/countframe['count'].sum()

    array_list=list(countframe['proportion'])
    print(array_list, "In")
    to_fill=np.random.choice(countframe['name'].values, size=(frame[column].isnull().sum(),), p=array_list)
    print(to_fill, "Inhere")
    print(len(isnull))
    frame.loc[isnull, column] = to_fill
    return frame
```

```python
loan_data=replace_with_proportion(loan_data, 'Married')
loan_data=replace_with_proportion(loan_data, 'Gender')
loan_data=replace_with_proportion(loan_data, 'Dependents')
```

```python
loan_data.columns[loan_data.isnull().any()]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
def factorize(data, var, fill_na = None):
      if fill_na is not None:
            data[var].fillna(fill_na, inplace=True)
      le.fit(data[var])
      data[var] = le.transform(data[var])
      return data
```

```python
numeric_data = loan_data.select_dtypes(include=[np.number])
cat_data = loan_data.select_dtypes(exclude=[np.number])
```

```python
loan_data.select_dtypes(exclude=[np.number]).columns
```

```python
loan_data = factorize(loan_data, "Gender")
loan_data = factorize(loan_data, "Married")
loan_data = factorize(loan_data, "Dependents")
loan_data = factorize(loan_data, "Education")
loan_data = factorize(loan_data, "Self_Employed")
loan_data = factorize(loan_data, "Property_Area")
```

```python
labels_train=labels_train.map({'Y': 1, 'N': 0})
```

```python
labels_train
```

```python
X_train=loan_data.iloc[0:614, :]
```

```python
X_test=loan_data.iloc[614:, :]
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
svc1=SVC(C=100, kernel='rbf')
svc1.fit(X_train, labels_train)
ypred=svc1.predict(X_test)
y_pred=pd.DataFrame(ypred)
y_pred=y_pred.iloc[:, 0].map({1: 'Y', 0: 'N'})
writer = pd.ExcelWriter('outputsvc1.xlsx')
y_pred.to_excel(writer,'Sheet1')
writer.save()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=25, min_samples_split=25, max_depth=7, max_features=5)
rfc.fit(X_train, labels_train)
ypred=rfc.predict(X_test)
y_pred=pd.DataFrame(ypred)
y_pred=y_pred.iloc[:, 0].map({1: 'Y', 0: 'N'})
writer = pd.ExcelWriter('outputrf1.xlsx')
y_pred.to_excel(writer,'Sheet1')
writer.save()
```
